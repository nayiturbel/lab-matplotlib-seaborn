#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.subplot(1,2,1) # index para ambos gráficos de 2
plt.plot(x,y,lw=1,color="grey", ls='--') #juego con los formatos y colores
plt.subplot(1,2,2)
plt.plot(x,z,lw=10, color="purple");
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig,axes=plt.subplots(1,4) # segunda forma sin index, primero filas, luego columnas.
#A diferencia del primer código, en este axes indica el núm de gráfico como tipo index
axes[0].plot(x,y,color="blue", lw=1, ls='-')
axes[1].plot(x,z,color="pink", lw=2, ls='-')
axes[2].plot(x,y,color="yellow", lw=3, ls='-')
axes[3].plot(x,z,color="purple", lw=4, ls='-');
```

#### Augmenting your previous code, resize your previous plot.

```python
# your code here
fig,axes=plt.subplots(nrows= 1,ncols= 4, figsize= (16,4)) # segunda forma sin index, primero filas, luego columnas.
#A diferencia del primer código, en este axes indica el núm de gráfico como tipo index
axes[0].plot(x,y,color="blue", lw=1, ls='-')
axes[1].plot(x,z,color="pink", lw=2, ls='-')
axes[2].plot(x,y,color="yellow", lw=3, ls='-')
axes[3].plot(x,z,color="purple", lw=4, ls='-');
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig,axes=plt.subplots(nrows= 1,ncols= 4, figsize= (16,4)) # segunda forma sin index, primero filas, luego columnas.
#A diferencia del primer código, en este axes indica el núm de gráfico como tipo index
axes[0].plot(x,y,color="blue", lw=1, ls='-')
axes[0].set_xlabel('Número de día')
axes[0].set_ylabel('Dinero')

axes[1].plot(x,z,color="pink", lw=2, ls='-')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

axes[2].plot(x,y,color="yellow", lw=3, ls='-')
axes[2].set_xlabel('Distancia')
axes[2].set_ylabel('Velocidad')

axes[3].plot(x,z,color="purple", lw=4, ls='-')
axes[3].set_xlabel('X')
axes[3].set_ylabel('Y');
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
# No entiendo si esto debería de ir en un solo gráfico o si son dos gráficos en uno como lo hice aquí
fig, axes = plt.subplots(1,2, figsize=(10,5))

labels=['y=x^2', 'y=e^x']

axes[0].plot(x,x**2,x, np.exp(x), color="purple", lw=2, ls='--')
axes[0].legend(labels)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title("Normal scale")

axes[1].plot(x,x**2,x, np.exp(x), color="purple", lw=2, ls='-.')
axes[1].legend(labels)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_yscale('log')
axes[1].set_title("Logarithmic scale (y)");
```
